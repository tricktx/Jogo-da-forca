```python
# Palavras que irão ser escolhidas aleatoriamente

palavras = [
    'Abacaxi',
    'Abobora',
    'Peixe',
    'Batata',
    'Queijo',
    'Zumbi',
    'Computador',
    'Mouse',
    'Teclado',
    'Fio',
    'Monitor',
    'Bola',
    'Microfone',
    'Livro',
    'Fone',
    'Telefone',
    'Flor',
    'Camera',
    'Suporte',
    'Espada',
    'Machado'
]
```

```python
# Importando as bibliotecas

import random
import time
```

```python
# Selecionando palavras
def selecionar_palavras():
    palavra = random.choice(palavras)
    return palavra.upper()



# Iniciar o jogo
def jogar(palavras):
    #Definindo algumas variáveis
    palavra_a_completar = "_" * len(palavras)
    advinhou = False
    letras_utilizadas = [ ]
    palavras_utilizadas = [ ]
    tentativas = 6



    #Boas vindas ao jogador! 
    print('Vamos jogar!!!')
    time.sleep(1.5)
    print(exibir_forca(tentativas))
    time.sleep(1.5)
    print("Esta é a palavra: %s" %palavra_a_completar)
    time.sleep(1.5)



    # Enquanto o usuário não advinhar e ainda houver tentativas
    while not advinhou and tentativas > 0:
        tentativa = input('Digite uma palavra ou letra para continuar: ').upper()
        time.sleep(1.5)
        if tentativa == int and tentativa == float:
            print('Não aceitamos números.')


        # Verificar se a letra digitada é única.
        if len(tentativa) == 1 and tentativa.isalpha():
        
        
            # Verificar se a letra já foi utilizada
            if tentativa in letras_utilizadas:
                print('A letra %s já foi utilizada. Tente novamente' %tentativa)
                time.sleep(1.5)


            # Verificar se a letra não está na palavra
            elif tentativa not in palavras:
                print("A letra %s não está na palavra" % tentativa)
                tentativas = tentativas - 1
                letras_utilizadas.append(tentativa)
                time.sleep(1.5)


            # Quando a Letra está na palavra
            else: 
                print('você acertou! A letra %s está na palavra!' %tentativa)
                letras_utilizadas.append(tentativa)
                time.sleep(1.5)



                # Transformando uma palavra em uma lista
                palavra_lista = list(palavra_a_completar)



                # Verificar onde podemos substituir o underline baseado na letra que digitamos
                indices = [i for i, letra in enumerate(palavras) if letra == tentativa]
                for indice in indices:
                    palavra_lista[indice] = tentativa
                palavra_a_completar = ''.join(palavra_lista)




                # Verificar condições de vitórias
                if "_" not in palavra_a_completar:
                    advinhou = True
        
        # Tentativa de acerta a palavra
        elif len(tentativa) == len(palavras) and tentativa.isalpha():



            # Palavra já utilizada
            if tentativa in palavras_utilizadas:
                print('A palavra %s já foi utilizada. Tente novamente' %tentativa)
                time.sleep(1.5)


                
            # Verificar se a palavra já não foi utilizada
            elif tentativa != palavras:
                print('Palavra %s digitada não é a correta' %tentativa)
                tentativas = tentativas - 1
                palavras_utilizadas.append(tentativa)
                time.sleep(1.5)



            # Acertou a palavra
            else:
                advinhou = True
                print('Você acertou a palavra %s.' %tentativa)
                palavra_a_completar = palavras
                time.sleep(1.5)
        


        # Tentativa inválida (números ou caracteres especiais)
        else: 
            print('Tentativa inválida, tente novamente!')
            tentativas = tentativas - 1
            time.sleep(1.5)



        # Status do jogo
        print(exibir_forca(tentativas))
        print(palavra_a_completar)

        # Finalizar o jogo caso o usuário acertou a palavra ou acabou as tentativas

    time.sleep(3)
    if advinhou:
        print
        print('Parabéns! Acertou a palavra!')
        time.sleep(1.5)
        
    else: 
        print('Acabaram as tentativas, a palavra era: %s' %palavras)
        time.sleep(1.5)
```

```python
def exibir_forca(tentativas):
  estagios = [  # Fim de jogo
              """
                  --------
                  |      |
                  |      O
                  |     \\|/
                  |      |
                  |     / \\
                  -
              """,
              # Falta 1 tentativa
              """
                  --------
                  |      |
                  |      O
                  |     \\|/
                  |      |
                  |     / 
                  -
              """,
              # Faltam 2 tentativas
              """
                  --------
                  |      |
                  |      O
                  |     \\|/
                  |      |
                  |      
                  -
              """,
              # Faltam 3 tentativas
              """
                  --------
                  |      |
                  |      O
                  |     \\|
                  |      |
                  |     
                  -
              """,
              # Faltam 4 tentativas
              """
                  --------
                  |      |
                  |      O
                  |      |
                  |      |
                  |     
                  -
              """,
              # Faltam 5 tentativas
              """
                  --------
                  |      |
                  |      O
                  |    
                  |      
                  |     
                  -
              """,
              # Estado inicial
              """
                  --------
                  |      |
                  |      
                  |    
                  |      
                  |     
                  -
              """
  ]

  return estagios[tentativas]
```

```python
def iniciar():
  palavra = selecionar_palavras()
  jogar(palavra)

  # Verificar se o usuário quer continuar
  while input('Deseja jogar novamente (S/N): ').upper() == 'S':
      palavra = selecionar_palavras()
      jogar(palavra)


iniciar()
```
